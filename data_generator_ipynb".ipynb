```python
import PyMieScatt as ps
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
import math
```

```python
env_refractive_index = 1.333
wavelength = 660  # nm wavelength
```

```python
theta, _, _, scatter_unpolarized = ps.ScatteringFunction(
                                        m = 1.1,  # particle refractive index
                                        wavelength = 660,
                                        diameter = 5.4,  # in nanometrs
                                        nMedium = 1.0,  # refractive index of enviroment
                                        minAngle = 10,
                                        maxAngle = 65,
                                        angularResolution = 55 / 256,
                                        space = 'theta',
                                        angleMeasure = 'degrees',
                                        normalization = None
                                    )
```

```python
database = pd.DataFrame(columns=['rel_r_i', 'size'] + list(theta))
database
```

```python
for i in range(20000):
  n = random.uniform(1.01,1.23)
  x = random.uniform(11,100)
  _, _, _, scatter_unpolarized = ps.ScatteringFunction(
                                        m = n*env_refractive_index,  # particle refractive index
                                        wavelength = 660,
                                        diameter = x/(1.333*math.pi/0.66)*1000,  # in nanometrs
                                        nMedium = env_refractive_index,  # refractive index of enviroment
                                        minAngle = 10,
                                        maxAngle = 65,
                                        angularResolution = 55 / 256,
                                        space = 'theta',
                                        angleMeasure = 'degrees',
                                        normalization = None
                                    )

  database.loc[i] = [n, x] + list(scatter_unpolarized)
```
